## Exercise 2

Let's create a ChatGPT of our own

```python
from dotenv import load_dotenv
import os
from openai import OpenAI
import pandas as pd

load_dotenv()  # Load variables from .env file
api_key = os.getenv("OPENAI_API_KEY")

# Create an OpenAI client instance
client = OpenAI(api_key=api_key)
L =[]
```

```python
instructions_prompt = "init"  # Initialize before the loop

while instructions_prompt.lower() != "exit":
    context_prompt = f"you are an assistance, your memory for the user chat is stored in {L} and you have to answer the user question based on that memory, if you don't have enough information to answer the question, you have to say that you don't have enough information to answer the question."
    instructions_prompt = input("Please enter your question: ")

    if instructions_prompt.lower() == "exit":
        break
    completion = client.chat.completions.create(
        model="gpt-4o",
        messages=[
        {"role": "system", "content":context_prompt},
        {"role": "user", "content": instructions_prompt}
        ]
    )

    L.append(completion.choices[0].message.content)
    print(f"User: {instructions_prompt}")
    print(f"Assistant: {completion.choices[0].message.content}")
```
