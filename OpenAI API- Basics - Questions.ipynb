## Exercise 3

### Use OpenAI API to add a new COLUMN for `df`

- Name the column `ProductSize`
- Categorize each product to be `Big` or `Small` base on the `QuantityPerUnit` column
<br></br>

- For each API call only ONE product `QuantityPerUnit` should be passed
    <p style="font-size:20px;color:Tomato"> USE a `for loop` to do so </p>

````python
df = pd.read_csv("NORTHWND_prd.csv", encoding="cp1252")
df = df.loc[df.CategoryID == 1, ['ProductID', 'ProductName', 'CategoryID','QuantityPerUnit', 'UnitPrice', 'UnitsInStock']]
df["ProductSize"]= None

context_prompt = "You are a skillful analyst that decides if Product is big or small based on the QuantityPerUnit column that you receive, analyze if that quantity per unit makes it Big or Small and you answer your result in a single word Big or Small."
for i in df.index:
  
  instructions_prompt = f"""
  Given the following QuantityPerUnit: {df.loc[i, "QuantityPerUnit"]}, make your analysis and decide if the Product is big or small and answer in a single word Big or Small.:
  
  """
  print(df.loc[i, "QuantityPerUnit"])
  completion = client.chat.completions.create(
    model="gpt-4o",
    messages=[
      {"role": "system", "content":context_prompt},
      {"role": "user", "content": instructions_prompt}
    ]
  )

  df.loc[i, "ProductSize"] = completion.choices[0].message.content
  print(df.loc[i, "ProductSize"])
df
````

## Exercise 6

### Use OpenAI API to create the COLUMN 'ProductSize' for `df`

with the <b><u>PARAMETERS</u></b> and the <b><u>CONTEXT</u></b> from above

<p style="font-size:20px;color:Tomato"> This time the Product should be categorize as "Big" or "Small" <u>  COMPARED  </u> to all other Products</p>

````python
df = pd.read_csv("NORTHWND_prd.csv", encoding="cp1252")
df = df.loc[df.CategoryID == 1, ['ProductID', 'ProductName', 'CategoryID','QuantityPerUnit', 'UnitPrice', 'UnitsInStock']]
df["ProductSize"] = None
all_quantities = df["QuantityPerUnit"].tolist()

context_prompt = "You're helping me to categorize each product by its 'QuantityPerUnit' description , the categories are 'Big' or 'Small'."
for i in df.index:
  
  instructions_prompt = f"""
   Given the following QuantityPerUnit: {df.loc[i, "QuantityPerUnit"]}, and all products have these QuantityPerUnit values: {all_quantities}.
  Make your analysis and decide if the Product is big or small in comparison to all the other products and return a single word: Big or Small.
  """
  completion = client.chat.completions.create(
    model="gpt-4o",
    messages=[
      {"role": "system", "content":context_prompt},
      {"role": "user", "content": instructions_prompt}
    ],
    temperature=0.0,
    max_tokens=5
  )
  df.loc[i, "ProductSize"] = completion.choices[0].message.content
  
df
````
