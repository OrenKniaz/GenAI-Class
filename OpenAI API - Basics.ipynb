## Roles

Roles in API Calls

- **System**
  - Provides context or instructions for the model
- **User**
  - Represents the end-user input
- **Assistant**
  - Model-generated responses
- **tool**
  - When you enable tool/function calling
- **function**
  - legacy - Same purpose as tool, accepted for backward compatibility.

>Scenario:
<br></br>We want the assistant to act as a professional python tutor helping with a question about JOIN/MERGE types.

````python
completion = client.chat.completions.create(
  model="gpt-4o",
  messages=[
        {"role": "system", "content": "You are a helpful Python tutor who gives clear and concise answers, keep the answers very short."},
        {"role": "user", "content": "Can you explain the difference between INNER JOIN and LEFT JOIN?"}
  ]
)

print(completion.choices[0].message.content)
````

But what happens if we'll ask a <u>follow up question</u> ??

````python
completion = client.chat.completions.create(
  model="gpt-4o",
  messages=[
        {"role": "user", "content": "Can you show an example with Python code?"}
  ]
)

print(completion.choices[0].message.content)
````

<br></br>
Hmmm.. Interesting, it doesn't know what we're asking for...
<br>
<p style="font-size:25px">Why is that ??</p>
<p style="font-size:30px"><u>Conversation Continuity</u></p>
The “memory” we
experience is recreated on the fly by re‑sending recent messages in the prompt.
<br>
<b>If nothing is passed back in, <u>the model starts from scratch</u></b>
<br></br>
<p style="font-size:25px">we'll use:</p>

```python
"role":"assistant"
```

````python
# First response:

completion = client.chat.completions.create(
  model="gpt-4o",
  messages=[
        {"role": "system", "content": "You are a helpful Python tutor who gives clear and concise answers, keep the answers very short."},
        {"role": "user", "content": "Can you explain the difference between INNER JOIN and LEFT JOIN?"}
  ]
)

print('', end='\n\n')
print('First response:')
print('----------------')
print(completion.choices[0].message.content)


# Save Assistant response
assistant_response = {"role":"assistant", "content":completion.choices[0].message.content}



# Second response:
completion = client.chat.completions.create(
  model="gpt-4o",
  messages=[
        {"role": "system", "content": "You are a helpful Python tutor who gives clear and concise answers, keep the answers very short."},
        {"role": "user", "content": "Can you explain the difference between INNER JOIN and LEFT JOIN?"},
        assistant_response,
        {"role": "user", "content": "Can you show an example with Python code?"}
  ]
)
print('', end='\n\n\n')

print('Second response:')
print('-----------------')
print(completion.choices[0].message.content)
````

### Audio and speech

There are multiple APIs for generating audio,
<br>
for example, we can do that with `/chat/completions` API or the `audio/speech` API (and there are a few others)
<br>
<p style="font-size:25px">So, what is the right one ??</p>

- <u>For general use APIs such as `/chat/completions`, you can:</u>
  - Use latest models' native audio understanding and generation capabilities
  - Combine them with other features like function calling
  - Select the model you want to use
  - **Depending on the control you need - Less control and predictability (the model thinks and responds in speech)**
  - The conversation will feel more natural

- <u>Specialized APIs such as `audio/speech`:</u>
  - Are specialized to work with specific models and only meant for one purpose
  - **More control and predictability - you know exactly what the model will say and can control the response**




<br>

#### Audio and speech - `/chat/completions`

````python
import base64

completion = client.chat.completions.create(
    model="gpt-4o-audio-preview",
    modalities=["text", "audio"],
    audio={"voice": "ash", "format": "wav"},
    messages=[
        {"role": "system", "content": "You are a helpful Python tutor who gives clear and concise answers, keep the answers very short."},
        {"role": "user", "content": "Can you explain the difference between INNER JOIN and LEFT JOIN in python?"}
    ]
)


# Save the audio response to a WAV file
wav_bytes = base64.b64decode(completion.choices[0].message.audio.data)
with open("tutor_chat_completions.wav", "wb") as f:
    f.write(wav_bytes)
````

<br>

#### Audio and speech - `audio/speech`

````python
response = client.audio.speech.create(
    model="gpt-4o-mini-tts",
    voice="fable",  # voices: alloy, echo, fable, onyx, nova, shimmer
    input="Hello! This is a test of OpenAI's text to speech API."
)

# Save the audio to a file
with open("speech_output.mp3", "wb") as f:
    f.write(response.content)
````

### Image generation & Vision

<br>

#### Image generation

<u>The Image generation API has 3 endpoints with different abilities:</u>

- **Generations**: Images from scratch, based on a text prompt
- **Edits**: Edited versions of images, where the model replaces some areas of a pre-existing image, based on a new text prompt
- **Variations**: Variations of an existing image

<br>

<u>Models:</u>

- **DALL-E 2**: Generations, edits, variations
- **DALL-E 3**: Only Generations - Higher quality

````python
import requests

response = client.images.generate(
    model="dall-e-3",
    prompt="a black sheep",
    size="1024x1024",
    quality="standard",
    n=1, # number of images to generate
)

print(response.data[0].url)

# Get the URL of the generated image
image_url = response.data[0].url

# Download the image
image_data = requests.get(image_url).content

# Save it to a file
with open("black_sheep.png", "wb") as f:
    f.write(image_data)
````

<br>

#### Vision

**Vision, does not have a separate endpoint — vision support is built into the `/chat/completions`**

<u>You can provide images as:</u>

- Qualified URL to an image file
- Image as a Base64-encoded data URL

````python
response = client.chat.completions.create(
    model="gpt-4o",
    messages=[
        {"role": "system", "content": "You are a helpful assistant, keep answers very short."},
        {
            "role": "user",
            "content": [
                {"type": "text", "text": "What is in this image?"},
                {
                    "type": "image_url",
                    "image_url": {
                        "url": image_url
                    }
                }
            ]
        }
    ]
)

print()
print(response.choices[0].message.content)
````

### Moderation

Identify potentially harmful content in text and images.

 **The Moderation endpoint is free to use.**

<br>

<u>The output has several categories in the JSON response:</u>

- **flagged**: `True` potentially harmful, `False` otherwise

- **categories**: `True` category it triggered, `False` otherwise

- **category_scores**: Confidence scores (between 0 and 1)

<br>

#### Moderate text

````python
response = client.moderations.create(
    model="omni-moderation-latest",
    input="I hate you and want to hurt you."
)

# Print the result
output = response.results[0]

print()
print("Flagged:")
print("--------")
print("Flagged:", output.flagged)

print(end='\n\n')
print("Categories:")
print("-----------")
display(output.categories.to_dict())


print(end='\n\n')
print("Categories scores:")
print("------------------")
display(output.category_scores.to_dict())
````

## API Parameters (temperature, max_tokens, top_p)

- **temperature**
  - Controls creativity (range 0 to 2)
  - Higher -> more random
  - Lower -> more focused and deterministic
  - **Generally recommend altering this or `top_p` but not both**
- **max_tokens**
  - The maximum tokens in the response
- **top_p**
  - Nucleus sampling; considers tokens from the top p fraction of probabilities
  - An alternative to sampling with temperature
  - Considers the results of the tokens with top_p probability mass.
  <br>So 0.1 means only the tokens comprising the top 10% probability mass are considered
  - **Generally recommend altering this or `temperature` but not both**

````python
# No parameters

completion = client.chat.completions.create(
  model="gpt-4o",
  messages=[
    {"role": "system", "content": "You are a poetic assistant, skilled in explaining complex programming concepts with creative flair."},
    {"role": "user", "content": "Compose a poem that explains the concept of recursion in programming."}
  ]
)

print(completion.choices[0].message.content)
````

````python
# temperature=0.1, max_tokens=50

completion = client.chat.completions.create(
    model="gpt-4o",
  messages=[
    {"role": "system", "content": "You are a poetic assistant, skilled in explaining complex programming concepts with creative flair."},
    {"role": "user", "content": "Compose a poem that explains the concept of recursion in programming."}
  ],
    temperature=0.1,
    #top_p=0.1,
    max_tokens=50
)

print(completion.choices[0].message.content)
````

````python
# top_p=0.1, max_tokens=50

completion = client.chat.completions.create(
    model="gpt-4o",
  messages=[
    {"role": "system", "content": "You are a poetic assistant, skilled in explaining complex programming concepts with creative flair."},
    {"role": "user", "content": "Compose a poem that explains the concept of recursion in programming."}
  ],
    #temperature=0.1,
    top_p=0.1,
    max_tokens=50
)

print(completion.choices[0].message.content)
````

<br>

<u>`max_tokens` Truncation</u>

When max_tokens Is Too Low

- The model stops generating once it hits max_tokens
- You may get an incomplete or cut-off response
- The `finish_reason` will be `"length"` in the `completion` object

````python
print("Finish reason:", completion.choices[0].finish_reason)
````

If you get `"length"` as the finish reason, and the output feels incomplete — you can:

- Increase `max_tokens`
- Use a follow-up prompt like `"continue"`
- Use streaming to dynamically detect truncation and react

## Tokens - Estimate number of tokens

- Tokens - smallest text chunk the model sees and bills for
- Approximately:   <u>≈ 4 characters</u>   or   <u>¾ word</u>
- Can be whole words, pieces, or punctuation

<br>
<u>Example: Sentence → Tokens</u>

`“ChatGPT is amazing.”` → `["Chat", "GPT", " is", " amazing", "."]`

So, 3-words, 22‑character = counts as 5 tokens

>Scenario
<br></br>
Let's say we want to estimate hoe many tokens are there in a doc


First, Create the doc `poem.txt`

````python
# Create a chat completion using the OpenAI client

completion = client.chat.completions.create(
  model="gpt-4o",
  messages=[
    {"role": "system", "content": "You are a poetic assistant, skilled in explaining complex programming concepts with creative flair."},
    {"role": "user", "content": "Compose a poem that explains the concept of recursion in programming."}
  ]
)

print(completion.choices[0].message.content)

# Save to file
with open("poem.txt", "w") as f:
    f.write(completion.choices[0].message.content)
````

Now, estimate number of tokens

````python
import tiktoken, pathlib
enc = tiktoken.encoding_for_model("gpt-4o-mini")
n_tokens = len(enc.encode(pathlib.Path("poem.txt").read_text()))
print(f' number od tokens: {n_tokens}')
````

`tiktoken`
library from OpenAI that splits text into the same byte‑pair tokens GPT models use

<u>Handy for:</u>

- Counting tokens before sending a prompt
- Previewing cost / truncation
- Chunking documents to fit a context window

### Second Step - API Call

 - Instructing the API to create a code for bar-chart

````python
# Define the prompt for the API call

context_prompt = "You are a helpful assistant that writes Python code."

instructions_prompt = f"""
Given the following DataFrame, generate Python code to create a visualization:
{df.to_string()}

Create a bar chart x-axis is CategoryID and the avg unitprice as y-axis.
The response should include only the Python code without any additional formatting or comments.

"""


completion = client.chat.completions.create(
  model="gpt-4o",
  messages=[
    {"role": "system", "content":context_prompt},
    {"role": "user", "content": instructions_prompt}
  ]
)

generatedCode = completion.choices[0].message.content

generatedCode = generatedCode.replace("```","")
generatedCode = generatedCode.replace("python","")
generatedCode


print(generatedCode)
````

````python
# Execute the generated code
exec(generatedCode)
````
